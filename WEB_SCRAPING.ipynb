```python
import requests
from bs4 import BeautifulSoup
import time

# Function to fetch and parse a single page
def fetch_page_data(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return BeautifulSoup(response.text, 'html.parser')
    else:
        return None

# Function to extract data from a page
def extract_data(soup):
    ratings = []
    identities = []
    Star_rating = []

    # Extract titles from span elements
    for item in soup.find_all('span', class_='analyst-rating-icon-neutral'):
        title = item.get('title', 'No title')
        ratings.append(title)
        
    # Extract text from a elements
    for tag in soup.find_all('a', class_='identity-cell-value'):
        text = tag.text.strip()
        identities.append(text)
    for item in soup.find_all('td' , class_ ='mds-data-table__cell'):
        title = item.get('title', 'No title')
        Star_rating.append(title)
    
    
    return ratings, identities, Star_rating

# Function to generate the URL for each page
def generate_url(base_url, page_number):
    return f"{base_url}?page={page_number}"

# Base URL of the website
base_url = 'https://www.morningstar.in/featured-reports.aspx'

# Number of pages to scrape
num_pages = 200

# Loop through each page and scrape data
all_ratings = []
all_identities = []
all_star_rating = []
for page in range(1, num_pages + 1):
    url = generate_url(base_url, page)
    print(f"Fetching data from {url}")
    soup = fetch_page_data(url)
    if soup:
        ratings, identities, Star_rating = extract_data(soup)
        all_ratings.extend(ratings)
        all_identities.extend(identities)
        all_star_rating.extend(Star_rating)
    time.sleep(1)  # Adding a delay to be polite and not overwhelm the server

# Print all the extracted data
print("Ratings:", all_ratings)
print("Identities:", all_identities)
print('Star_rating',all_star_rating )
```
